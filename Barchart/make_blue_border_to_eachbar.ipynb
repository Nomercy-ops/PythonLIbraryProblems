@Title :  Program  to display a bar chart of the popularity of programming
Languages. Make blue border to each bar.


```python
from matplotlib import pyplot as plt
from LogHandler import logger
```

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
def make_blue_border():
   try:
      plt.bar(x,y,fc ='orange',edgecolor='blue')
      plt.xlabel("Programming Languages") 
      plt.ylabel("Popularity")   
      plt.title("Popularity of Programming Language")
      plt.ylim(0,25)
      plt.minorticks_on()
      plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
      # Customize the minor grid
      plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

   except Exception as e:
      logger.error(e)

make_blue_border()
```

@Title :  Program  to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.


```python
def specify_position():
   try:
      x_pos = [i for i, _ in enumerate(x)]
      plt.xlabel("Languages")
      plt.ylabel("Popularity")
      plt.title("Popularity of Programming Language")
      plt.xticks(x_pos, x,rotation = 'vertical' )
      
      # Select the position of each barplots on the x-axis (space=1,3,3,2,1)
      y_pos = [0,1,3,5,7,9]
      # Create bars
      plt.bar(y_pos, y,color = 'orange',width = 0.55)
      plt.xticks(y_pos, x)
      plt.ylim(0,25)
      
      plt.minorticks_on()
      plt.grid(which='major', linestyle='-', linewidth='0.5', color='blue')
      # Customize the minor grid
      plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

   except Exception as e:
      logger.error(e)

specify_position()
```

@Title :  Program  to display a bar chart of the popularity of programming
Languages. Select the width of each bar and their positions.


```python
def select_width_for_each_bar():
   try:
        x_pos = [i for i, _ in enumerate(x)]
        plt.xlabel("Languages")
        plt.ylabel("Popularity")
        plt.title("Popularity of Programming Language")
        plt.xticks(x_pos, x)
        
        # Select the width of each bar and their positions
        width = [0.5,0.4,0.6,1.1,0.7,0.6]
        y_pos = [0,1.5,3,5,7,9]
        # Create bars
        plt.bar(y_pos, y,color = 'orange',width = width)
        plt.xticks(y_pos, x)
        plt.ylim(0,25)
        
        plt.minorticks_on()
        plt.grid(which='major', linestyle='-', linewidth='0.5', color='blue')
        # Customize the minor grid
        plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

   except Exception as e:
      logger.error(e)

select_width_for_each_bar()
```
