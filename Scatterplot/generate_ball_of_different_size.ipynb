@Title : Program to draw a scatter plot using random distributions to generate balls of
different sizes.

```python
from matplotlib import pyplot as plt
import math
import pandas as pd 
import numpy as np
from LogHandler import logger

try:
    x = np.array(np.random.rand(20))
    y = pd.Series(np.random.rand(20))
    colors = [np.random.randint(1, 4) for i in range(20)]
    areas = [math.pi * np.random.randint(5, 15)**2 for i in range(20)]
    plt.figure(figsize=(9,7))
    plt.scatter(x, y, s=areas, c= colors,alpha=0.7)
    plt.xlabel("X")
    plt.ylabel("Y")

except Exception as e:
    logger.error(e)
```
