@Title : Program to  draw a scatter plot comparing two subject marks of Mathematics
and Science. Use marks of 10 students.

```python
from matplotlib import pyplot as plt
import pandas as pd
from LogHandler import logger

try:
    df = pd.read_csv('fdata.csv')
    marks_range = df['range']
    math_marks = df['math']
    science_marks = df['science']
    
    plt.scatter(marks_range, math_marks,s=50, label='Math marks', color='r')
    plt.scatter(marks_range, science_marks,s=50, label='Science marks', color='g')

    plt.title('Scatter Plot')
    plt.xlabel('Marks Range')
    plt.ylabel('Marks Scored')
    plt.legend()

except Exception as e:
    logger.error(e)
```
