@Title : Program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other.

```python
from matplotlib import pyplot as plt
import numpy as np
from LogHandler import logger

try:
    x = np.random.randint(0,50,size=(15, 1))
    y = np.random.randint(0,50,size=(15, 1))
    plt.scatter(x, y, s=60, facecolors='none', edgecolors='k')
    plt.xlabel("X")
    plt.ylabel("Y")

except Exception as e:
    logger.error(e)
```
