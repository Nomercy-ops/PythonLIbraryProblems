@Title : Program to create a pie chart with a title of the popularity of programming
Languages.Make multiple wedges of the pie.

```python
from matplotlib import pyplot as plt 
import pandas as pd
from LogHandler import logger 

try:
    df =  pd.read_csv('fdata.csv')
    lang_data = df["language"]
    popular_data = df["popularity"]
    fig = plt.figure(figsize=(9,6))
    fig.patch.set_facecolor('w')
    # Plot
    explode = (0.1, 0, 0, 0, 0, .1)  
    plt.pie(popular_data, explode=explode, labels=lang_data,
    autopct='%1.1f%%', startangle=90)
    plt.title('popularity of programming Languages',bbox={'facecolor':'0.8', 'pad':4})

except Exception as e:
    logger.error(e)
```
