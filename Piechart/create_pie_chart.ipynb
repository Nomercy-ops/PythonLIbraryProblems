@Title : Program to create a pie chart of the popularity of programming
Languages.

```python
from matplotlib import pyplot as plt 
from LogHandler import logger 
import pandas as pd

try:
    
    df =  pd.read_csv('fdata.csv')
    # df = df.astype({"popularity": int})
    lang_data = df["language"]
    popular_data = df["popularity"]

    fig = plt.figure(figsize=(5,5))
    fig.patch.set_facecolor('w')
    # Plot
    plt.pie(popular_data,labels=lang_data,autopct='%1.1f%%',startangle=90)

except Exception as e:
    logger.error(e)
```
