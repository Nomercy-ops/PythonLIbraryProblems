@Title :  Program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.


```python
from matplotlib import pyplot as plt
from LogHandler import logger

try:
    with open("test.txt") as f:
        data = f.read()
    data = data.split('\n')
    x = [float(row.split(' ')[0]) for row in data]
    y = [float(row.split(' ')[1]) for row in data]
    plt.plot(x,y)
    plt.xlabel('x - axis')
    plt.ylabel('y - axis')
    plt.title('Sample Graph')
    plt.show()

except Exception as e:
    logger.error(e)
```
